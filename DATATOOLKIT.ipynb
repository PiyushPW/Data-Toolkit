6. What are the differences between NumPy arrays and Python lists?
  - 1. Performance

    NumPy Arrays: Faster due to their fixed type and optimized C implementation.

    Python Lists: Slower as they are more flexible and store elements as objects, adding overhead.
    2. Data Type

    NumPy Arrays: Homogeneous; all elements must be of the same data type.

    Python Lists: Heterogeneous; can store elements of different data types.
    3. Memory Efficiency

    NumPy Arrays: More memory-efficient because they store data in contiguous blocks of memory.

    Python Lists: Less memory-efficient as they store references to objects, which adds overhead.
    4. Functionality

    NumPy Arrays: Provide a wide range of mathematical, statistical, and linear algebra operations directly.

    Python Lists: Limited built-in operations; require loops or external libraries for complex computations.
    5. Indexing and Slicing

    NumPy Arrays: Support advanced slicing, boolean indexing, and multidimensional indexing.

    Python Lists: Support basic slicing but lack advanced indexing features.
    6. Mutability

    NumPy Arrays: Mutable; elements can be changed, but resizing is less straightforward.

    Python Lists: Mutable; elements can be changed, and resizing (adding/removing elements) is simple.

17. Explain the difference between apply() and map() in Pandas?
  - 1. map:

    Scope: Works only on a Pandas Series (1D data).

    Purpose: Applies a function element-wise to each value in the Series.

    Input: Can take a function, dictionary, or Series to map values.

    Output: Returns a transformed Series.

  - 2. apply:

    Scope: Works on both Series and DataFrames (1D or 2D data).

    Purpose: Applies a function along an axis (rows or columns for DataFrames).

    Input: Takes a function as input.

    Output: Returns a transformed Series (for Series input) or DataFrame (for DataFrame input).

```python
#4
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6],
    'D': [10, 20, 30, 40, 50]
}
df = pd.DataFrame(data)

# Step 1: Calculate the correlation matrix
correlation_matrix = df.corr()

# Step 2: Visualize the correlation matrix with a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#5
import plotly.graph_objects as go

# Data for the bar plot
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [10, 15, 7, 12]

# Create the bar plot
fig = go.Figure(data=[
    go.Bar(x=categories, y=values, marker_color='skyblue', name='Values')
])

# Customize layout
fig.update_layout(
    title='Bar Plot Example',
    xaxis_title='Categories',
    yaxis_title='Values',
    template='plotly_white'
)

# Show the plot
fig.show()
```

```python
#8
import matplotlib.pyplot as plt

# Data for the plot
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]  # Line 1
y2 = [2, 4, 6, 8, 10]   # Line 2
y3 = [5, 7, 9, 11, 13]  # Line 3

# Create the plot
plt.figure(figsize=(8, 6))  # Set the figure size

# Plot multiple lines
plt.plot(x, y1, label='Line 1: Squares', color='blue', linestyle='-', marker='o')
plt.plot(x, y2, label='Line 2: Doubles', color='green', linestyle='--', marker='s')
plt.plot(x, y3, label='Line 3: Incremental', color='red', linestyle='-.', marker='^')

# Add labels, title, and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Multiple Lines')
plt.legend()

# Show grid and display the plot
plt.grid(True)
plt.show()
```

```python
#13
import plotly.graph_objects as go
import numpy as np

# Generate random data
np.random.seed(42)
x = np.random.randn(50)
y = np.random.randn(50)
z = np.random.randn(50)

# Create a 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=5,
        color=z,  # Color by z-axis values
        colorscale='Viridis',  # Color scale
        opacity=0.8
    )
)])

# Add title and labels
fig.update_layout(
    title='3D Scatter Plot',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)

# Show the plot
fig.show()

```
